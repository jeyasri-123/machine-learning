```python
import pandas as pd
import numpy as np
```

```python
wine_data=pd.read_csv("C:\\Users\\msiva\\Downloads\\winequality-white.csv",
                      names=("Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides","Free Sulphur Dioxide","Total Sulphur Dioxide",
                             "Density","pH","Sulphates","Alcohol","Quality"),
                      skiprows=1,sep=r'\s*;\s*',engine='python')
wine_data.head()
#we use skiprows to avoid 1st row in the data as it is confusing
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corrmat=wine_data.corr()
f, ax=plt.subplots(figsize=(7, 7))
sns.heatmap(corrmat, vmax=.8, square=True,annot=True, fmt='.2f')
plt.show()
```

```python
x = wine_data.drop('Quality',axis=1)
y = wine_data['Quality']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2 ,random_state=0)
#here we say 20 percent of our data set used for test service
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=5,max_depth=10,max_features=5)
print(classifier.get_params())
#criterion='entrophy' by this increase the IG
```

```python
classifier.fit(x_train, y_train) 
```

```python
score = classifier.score(x_test, y_test)
print(score)
#after indicating estimator(parameter) the score changed from 62 to 50%
#we add parametera to make less complex and less accuracy
#each parameters working differ with the data set we use
```
